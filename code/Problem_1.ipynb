## Problem 1: A Quantum Coin Flip

The pauli-X gate is a deterministic gate, where the new coin state is just one minus the previous coin state. So basically we go from ∣x=0⟩⊗∣0⟩ to ∣x=1⟩⊗∣1⟩ to ∣x=0⟩⊗∣0⟩ and the cycle repeats. Thus, the path we get is 0,1,0,1,0,1... which is a simple, familiar and deterministic oscillation.

```python
position=0                 #as initial position is 0
coin_state=0               #initially the coin is in |0>
steps=5

print(position, end=", ")             #starting position

for i in range(steps):
    coin_state=1-coin_state           #coin operation
    
    if coin_state==0:                 #shift operation
        position-=1 
    else:
        position+=1 
    
    print(position,end=", ")
```
